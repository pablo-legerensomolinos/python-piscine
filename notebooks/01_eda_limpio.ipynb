# Calidad del aire en Madrid — análisis exploratorio

**Pregunta:** ¿cuánto contribuye el tráfico al NO₂ que respiramos en Madrid?

Este notebook usa el paquete `airemad` (en `src/`): aquí solo hay
**narrativa y resultados**. La lógica de carga, limpieza y gráficos
vive en módulos versionados y testeados.

```python
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt

from airemad import load_raw, limpiar
from airemad import plots
```

## 1. Carga y limpieza

Una línea. Toda la lógica está en `airemad.clean`, con tests.

```python
df = limpiar(load_raw())
print(f"{len(df)} filas | {df['ESTACION'].nunique()} estaciones | nulos NO2: {df['NO2'].isna().sum()}")
df.head()
```

## 2. Evolución mensual del NO₂

Dos cosas que buscar en el gráfico:
- **Estacionalidad**: el invierno es peor (calefacciones + inversión térmica).
- **Jerarquía de estaciones**: las de tráfico (Plaza Elíptica) muy por encima de las de parque (Casa de Campo).

```python
ax = plots.serie_mensual_no2(df)
plots.guardar(ax.figure, "no2_mensual", carpeta="../figures")
plt.show()
```

## 3. El experimento natural: el fin de semana

Cada sábado y domingo, Madrid hace un experimento sin querer: quita buena parte del tráfico
y deja todo lo demás igual. Si el NO₂ cae el finde, esa caída es ~atribuible al tráfico.

```python
ax = plots.comparativa_semana_finde(df)
plots.guardar(ax.figure, "no2_finde", carpeta="../figures")
plt.show()
```

## 4. Conclusiones

1. Las estaciones de tráfico intenso duplican con creces los niveles de las estaciones de fondo (parques).
2. El NO₂ cae de forma clara los fines de semana, y cae **más** donde más tráfico hay: la huella del coche es directa.
3. Ninguna estación cumple la guía OMS (10 µg/m³); las de tráfico rondan o superan el límite UE (40).
